```python
depressed=["rejects","bullied","lonely","dead","wrong"]
casual = ["excited","happy","party","good"]
index=0
for i in a["Sentences"]:
    for j in depressed:
        if j in i:
            a["Mood"].iloc[index]="depressed"
        else:
            pass
    for k in casual:
        if k in i:
            a["Mood"].iloc[index]="casual"
        else:
            pass
    index=index+1
```
